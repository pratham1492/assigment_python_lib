```python
# Q3 

import pandas as pd
import numpy as np

df = pd.DataFrame({
    "player":["Virat","Rohit","Dhoni","Gill","Hardik"],
    "runs":[800,np.nan,450,890,np.nan]
})

df["runs"] = df["runs"].fillna(df["runs"].mean())

df["runs"]
```

```python
# Q4

import pandas as pd
import numpy as np

df = pd.DataFrame({
    "restaurant":["A","B","C","D","E"],
    "rating":[4.5,np.nan,np.nan,4.2,3.8]
})

df

df["rating"] = df["rating"].ffill()

df["rating"] = df["rating"].bfill()

df
```

```python
# Q5

"ANSWER = The missing values in the runs column are most likely MCAR (Missing Completely At Random). The missing values appear randomly and do not seem to be related to any other column such as player name or matches played. Therefore, MCAR is the most suitable missingness mechanism for this dataset."
```
