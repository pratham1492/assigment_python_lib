```python
# Q1

import pandas as pd

df = pd.DataFrame({
    "restaurant": ["Dominos", "KFC", "Dominos", "Pizza Hut"],
    "date": ["2025-06-01", "2025-06-02", "2025-06-01", "2025-06-03"]
})

duplicates = df[df.duplicated(subset=["restaurant", "date"])]

print(duplicates)
```

```python
# Q4

import pandas as pd

df = pd.DataFrame({
    "username": [
        "insta_queen",
        "insta-queen",
        "instaqueen",
        "insta_queen"
    ]
})

df["username"] = df["username"].replace({
    "insta_queen": "instaqueen",
    "insta-queen": "instaqueen"
})

df
```

```python
# Q5

import pandas as pd

df = pd.DataFrame({
    "status": ["Yes", " yes ", "Y", "No", "no", "N", " Yes "]
})

df["status"] = df["status"].str.strip().str.lower()

df["status"] = df["status"].replace({
    "yes": 1,
    "y": 1,
    "no": 0,
    "n": 0
})

df
```
