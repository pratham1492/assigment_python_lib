```python
# Q1

import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

delivery = np.random.randint(20, 60, 50)

sns.histplot(delivery, bins=10)

plt.title("Zomato Delivery Time")
plt.xlabel("Minutes")
plt.ylabel("Count")
```

```python
# Q2

import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")

sns.boxplot(x="day", y="total_bill", data=tips)

plt.title("Total Bill by Day")
```

```python
# Q3

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

sns.set_theme(style="darkgrid")

teams = ["CSK","MI","RCB","GT","RR","LSG","SRH","KKR"]

team = np.repeat(teams, 10)
runs = np.random.randint(120, 230, 80)

df = pd.DataFrame({
    "Team": team,
    "Runs": runs
})

sns.violinplot(x="Team", y="Runs", data=df)

plt.title("IPL Team Scores")
```

```python
# Q4

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

genre = [
"Pop","Rock","HipHop","Classical",
"Pop","Rock","HipHop","Pop",
"Rock","Pop","Classical","HipHop",
"Pop","Rock","Pop","HipHop",
"Classical","Rock","Pop","HipHop",
"Pop","Rock","Classical","Pop",
"HipHop","Rock","Pop","Classical",
"Pop","HipHop","Rock","Pop",
"Rock","Classical","HipHop","Pop",
"Rock","Pop","HipHop","Classical"
]

df = pd.DataFrame({"Genre": genre})

sns.countplot(x="Genre", data=df)

plt.title("Spotify Song Genres")
```

```python
# Q5

import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")

steps = [4500,5200,6100,7000,6500,9000,8500]

sns.kdeplot(steps, fill=True, color="green")

plt.title("Daily Step Count")
```
