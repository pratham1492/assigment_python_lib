```python
# Q1

import matplotlib.pyplot as plt

fig, ax = plt.subplots(2, 2, figsize=(8, 8))

x = [1, 2, 3, 4]
y = [10, 20, 15, 25]

ax[0,0].plot(x, y)
ax[0,0].set_title("Line Plot")

apps = ["A", "B", "C"]
orders = [15, 20, 10]

ax[0,1].bar(apps, orders)
ax[0,1].set_title("Bar Chart")

rating = [4.1, 4.3, 4.5, 4.7]
price = [200, 250, 300, 350]

ax[1,0].scatter(rating, price)
ax[1,0].set_title("Scatter Plot")

sizes = [40, 30, 20, 10]
labels = ["A", "B", "C", "D"]

ax[1,1].pie(sizes, labels=labels, autopct="%1.1f%%")
ax[1,1].set_title("Pie Chart")

plt.tight_layout()
```

```python
# Q2

import matplotlib.pyplot as plt

apps = ["Zomato", "Swiggy", "Domino's"]
time = [30, 25, 40]

plt.bar(apps, time, color=["red", "orange", "blue"], linewidth=2)

plt.title("Average Delivery Time")
plt.xlabel("Apps")
plt.ylabel("Minutes")
```

```python
# Q3

import matplotlib.pyplot as plt

names = ["A", "B", "C", "D", "E"]
followers = [1200, 2500, 4000, 6000, 8000]
posts = [2, 3, 1, 4, 5]

fig, ax1 = plt.subplots()

ax1.plot(names, followers, color="blue", marker="o")
ax1.set_xlabel("Influencers")
ax1.set_ylabel("Followers", color="blue")

ax2 = ax1.twinx()

ax2.plot(names, posts, color="red", marker="s")
ax2.set_ylabel("Daily Posts", color="red")

plt.title("Followers and Daily Posts")
```

```python
# Q4

import matplotlib.pyplot as plt

movies = ["Movie1", "Movie2", "Movie3", "Movie4", "Movie5"]
ratings = [8.1, 7.5, 8.8, 7.9, 8.4]
tickets = [120, 95, 180, 110, 150]

plt.scatter(ratings, tickets)

for i in range(len(movies)):
    plt.annotate(movies[i], (ratings[i], tickets[i]))

plt.title("Movies vs Tickets Sold")
plt.xlabel("IMDB Rating")
plt.ylabel("Tickets Sold")
```

```python
# Q5

import matplotlib.pyplot as plt

months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]
sales = [120, 150, 180, 300, 220, 200]

plt.plot(months, sales, marker="o")

plt.annotate(
    "Big Billion Days",
    xy=("Apr", 300),
    xytext=("Mar", 330),
    arrowprops=dict(arrowstyle="->")
)

plt.title("Flipkart Monthly Sales")
plt.xlabel("Months")
plt.ylabel("Sales")
```
