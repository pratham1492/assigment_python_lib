```python
# Q3 

import pandas as pd

df = pd.DataFrame({
    "transaction_amount":[100,200,300,400,500,1000,5000]
})

low = df["transaction_amount"].quantile(0.05)

high = df["transaction_amount"].quantile(0.95)

df["transaction_amount"] = df["transaction_amount"].clip(low, high)

df["transaction_amount"].describe()
```

```python
# Q4

import pandas as pd

df = pd.DataFrame({
    "price":["₹1,299","₹2,499","₹999","₹3,599"]
})

df["price"] = df["price"].str.replace("₹","",regex=False)

df["price"] = df["price"].str.replace(",","",regex=False)

df["price"] = df["price"].astype(int)

df
```

```python
# Q5

import pandas as pd

df = pd.DataFrame({
    "is_premium":[True,"True",1,"yes",False,0,"no"]
})

df["is_premium"] = df["is_premium"].apply(
    lambda x: str(x).lower() in ["true","1","yes"]
)

df
```
