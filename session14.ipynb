```python
# Q1

import pandas as pd

df = pd.read_csv("deliveries.csv")

print(df.head())

print("\nSummary Statistics for total_runs")

print("Mean   :", df["total_runs"].mean())
print("Median :", df["total_runs"].median())
print("Min    :", df["total_runs"].min())
print("Max    :", df["total_runs"].max())
print("Std    :", df["total_runs"].std())
```

```python
# Q2

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("Dataset-SA.csv")

sns.countplot(x="rating", data=df)

plt.title("Count of Reviews by Rating")
plt.xlabel("Rating")
plt.ylabel("Number of Reviews")

plt.show()
```
