```python
# Q1

import matplotlib.pyplot as plt

days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
steps = [5000, 6200, 7000, 6500, 8000, 9500, 10000]

plt.plot(days, steps)

plt.title("Daily Steps")
plt.xlabel("Days")
plt.ylabel("Steps")

plt.savefig("steps_lineplot.png")
```

```python
# Q2

import matplotlib.pyplot as plt

ratings = [4.5, 4.2, 3.9, 4.7, 4.1, 3.8, 4.4, 4.0, 4.6, 4.3]
price = [300, 250, 180, 450, 220, 150, 350, 200, 400, 280]

plt.scatter(ratings, price)

plt.title("Restaurant Ratings vs Meal Price")
plt.xlabel("Zomato Rating")
plt.ylabel("Average Meal Price")
```

```python
# Q3

import matplotlib.pyplot as plt

apps = ["Swiggy", "Zomato", "Domino's"]
orders = [12, 18, 6]
colors = ["orange", "red", "blue"]

plt.bar(apps, orders, color=colors, label="Orders")

plt.title("Food Orders")
plt.xlabel("Apps")
plt.ylabel("Orders")
plt.legend()
```

```python
# Q4

import matplotlib.pyplot as plt

sessions = [25,30,45,60,35,20,40,55,50,65,70,30,25,45,60,40,35,50,55,30]

plt.hist(sessions, bins=5, color="green")

plt.title("Spotify Listening Sessions")
plt.xlabel("Minutes")
plt.ylabel("Frequency")
```

```python
# Q5

import matplotlib.pyplot as plt

days = ["Mon","Tue","Wed","Thu","Fri","Sat","Sun"]
screen = [2,3,2.5,4,3.5,5,4.5]
likes = [30,45,35,50,60,70,55]

fig, ax = plt.subplots(1,2,figsize=(10,4))

ax[0].plot(days, screen)
ax[0].set_title("Instagram Screen Time")
ax[0].set_xlabel("Days")
ax[0].set_ylabel("Hours")

ax[1].bar(days, likes, color="purple")
ax[1].set_title("Posts Liked")
ax[1].set_xlabel("Days")
ax[1].set_ylabel("Likes")

plt.savefig("social_media_usage.png")
```
